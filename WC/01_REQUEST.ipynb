```python
# 크롤링이란 우리가 할 작업을 컴퓨터에게 대신 시킨다.
# 네이버 홈페이지 가져오라고하면 브라우저 네이버 서버를 검색한다. https:// (브라우저 사용) = request
# 서버에서는 html의 형태를 Response 해준다. 
# 브라우저는 html을 해석해서 우리에게 보여준다.
# 우리는 html 코드속에서 필요한 데이터를 찾아내야한다.
```

```python
# requeists = 페이지 정보를 요청하는 라이브러리
# import작업 진행을 해야한다!
# 브라우저의 역할을 대신한다.
import requests as req
```

```python
# 나 : req야 네이버 메인페이지 정보 좀 알아줘!
# req : 네, 근데 네이버가 어디에요?
# 나 : www.naver.com
res = req.get("https://www.naver.com")
```

```python
# <Response [200]> = 데이터를 잘 가지고 왔습니다. 통신에 성공했습니다!
# 나 : res 변수에서 내용만 보고싶어 -> 브라우저가 없기 때문에 request 코드만 긁어온는거다!
res.text
```
