```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
#1. 네이버 금융사이트에서 코스피 탑100 사이트의 정보를 요청
res = req.get("https://finance.naver.com/sise/sise_quant.naver") # 변수에 집어넣기 전에 잘 들어왔나 꼭 확인해봐야한다.
```

```python
res.text
```

```python
#2. 사용자가 받아온 데이터 (text) 컴퓨터가 알아 먹을 수 있게 컴퓨터용 언어로 변형!
# html언어로 변형 > bs 라이브러리
```

```python
soup = bs(res.text,"lxml")
```

```python
#3. soup데이터에서 내가 필요한 요소를 찾아오기!
title = soup.select('a.tltle')
```

```python
price = soup.select('td.number:nth-child(3)') 
#tag 마우스 우클릭 -> copy -> copy selector
#contentarea > div.box_type_l > table > tbody > tr:nth-child(3) > td:nth-child(3))
```

## 거래량

```python
volume = soup.select("td.number:nth-child(6)")
```

```python
len(volume)
```

```python
for i in title :
    print(i.text)
```

```python
for i in price :
    print(i.text)
```

```python
for i in volume :
    print(i.text)
```

```python
# title,volume,price 리스트에서 순수한 글자 정보만 추출한 리스트를 새로 제작
titleList = []
priceList = []
volumeList = []

#웬만하면 list는 만들어서 초기화를 동시에 진행해서 error를 최소화하자!

for i in range(len(title)):
    titleList.append(title[i].text)
    priceList.append(price[i].text)
    volumeList.append(volume[i].text)
    
```

```python
volumeList # 사용자를 위해서 순수하게 글자 정보만 꺼내는 작업 -> list로 만들어 놓는 것 : Data setting
```

# 정제된(순수한 글자) 데이터를 표 형식 제작

```python
# Pandas, DataFrame
import pandas as pd
```

```python
data = {"종목명" : titleList, "현재가" : priceList, "거래량" : volumeList}
```

```python
kospi = pd.DataFrame(data) # 현재 표의 value 값만 가지고 있음
```

```python
#csv 파일로 만드는 방법
kospi.to_csv("코스피데이터.csv", encoding = "euc-kr")
# 기본적으로 한글을 잘 인식하지 않는다. 따라서 encoding 작업이 반드시 필요하다!
# 1. euc-kr 2. "" 3. utf-8-sig
```
