# 벅스 차트 TOP 100

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

### 1. 벅스차트 정보를 요청

```python
res = req.get('https://music.bugs.co.kr/chart?wl_ref=M_left_02_01')
```

### 2. 넘겨받은 데이터가 문자 형태이기 때문에
#### 컴퓨터가 알아먹을 수 있는 언어로 변형

```python
res.text
```

```python
soup = bs(res.text, 'lxml')
```

### 3. 순위,곡명,가수명 수집

```python
soup.select('strong') # 필요하지 않는 데이터도 같이 온다! - 부모 tag도 확인
```

```python
rank = soup.select('div.ranking > strong') # 구분자가 존재하지 않으면 부모를 확인하자
```

```python
title = soup.select('p.title > a')
```

```python
singer = soup.select('p.artist > a:nth-child(1)')
```

```python
len(rank)
```

```python
len(title)
```

```python
len(singer) # 경서예지처럼 가수가 2개인 경우도 있었네..? 페이지 구조에서 확인하자!
```

### # 4. 순수한 글자 정보만 담아줄 리스트 제작 (tag정보는 필요 없기 떄문)

```python
rankList=[]
titleList=[]
singerList=[]

for i in range(len(rank)):
    rankList.append(rank[i].text)
    titleList.append(title[i].text)
    singerList.append(singer[i].text)
```

```python
singerList
```

### 5. 순수한 글자정보를 가지고 데이터프레임(표) > 딕셔너리데이터를 제작

```python
data ={"순위" : rankList, "제목" : titleList, '가수' : singerList}
bugs = pd.DataFrame(data)
```

```python
# 인덱스 번호 꼴불견이네
bugs.set_index('순위')
```

### 6. csv 파일로 저장, html파일로 저장

```python
bugs.to_csv('bugs_chart.csv',encoding = 'euc-kr')
```

### html로 만들어보기

```python
bugs.to_html('bugs_chart.html')
```
