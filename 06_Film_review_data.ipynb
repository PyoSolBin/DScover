```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res = req.get('https://movie.naver.com//movie/bi/mi/pointWriteFormList.naver?code=208077&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=1')
```

```python
soup = bs(res.text,'lxml')
```

### 데이터가 수집되지 않는 경우는 서버가 다른지 의심해봐야한다!

```python
# 영화 관람객들의 리뷰 데이터를 수집하고자 한다. #부모의 부모까지 이용할 줄 알아야한다.
# id는 고유한 값 하나만 가져올 때 쓴다. class값은 여러가지 값을 중복으로 가져올때 사용한다.
# iframe 보이기만 할 뿐 해당 페이지에 데이터가 저장되어있지 않는 경우도 있다!!
# 데이터가 너무 방대하기 때문에 어떤 데이터들은 다른 서버에 저장해놓고 보여주기만 하는 경우가 있다.
review = soup.select("div.score_reple > p")
```

### iframe 복습
- 평범함 url ->
- response 200 -> 
- 실제 내가 원하는 데이터 x -> 
- 보여지기만 하는 데이터였던것임 -> 
- real data는 다른 server에 있었음 -> 
- tag에 접근하기 전에 알 수 있는 방법도 존재 -> 
- 마우스 우클릭 ->  
- (프레임 소스보기) : 실제 데이터가 저장되어 있는 공간이 따로 존재한다는 뜻임 -> 
- 검사 -> (ctrl + f) iframe(검색) ->
- 찾은 URL 서버 뒤에 붙여넣기 ->
- 실제 데이터의 서버를 확인할 수 있다!

```python
# 불필요한 개행문자(enter) 삭제하는 방법 strip()
for i in review:
    print(i.text.strip()) # 필요하지 않는 data 관람객 이라는 단어가 계속해서 등장 + 공백이 너무 많다
```

```python
for i in tq(range(1,30)):
    res = req.get('https://movie.naver.com//movie/bi/mi/pointWriteFormList.naver?code=208077&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='+str(i))
    soup = bs(res.text,'lxml')
    viewer = soup.select("span.ico_viewer")
    spo = soup.select('span.text_spo._text_spo')
    for j in viewer :
        j.extract()
    for p in spo:
        p.extract() 
    for i in review:
        print(i.text.strip())
```
